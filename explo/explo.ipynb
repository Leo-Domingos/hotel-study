```python
import pandas as pd

df = pd.read_csv('../data/hotel.csv')
df.head()
```

```python
df.describe().T
```

```python
df_booked = df.loc[df.booking_status == 'Not_Canceled'].reset_index(drop = True)
df_booked
```

```python
df_booked.loc[(df_booked.arrival_month == 2) & (df_booked.arrival_date == 29),'arrival_date'] = 28
```

```python
df_booked.loc[:, 'data_chegada'] = pd.to_datetime(df_booked.loc[:,'arrival_date'].astype(str)+'/'+df_booked.loc[:,'arrival_month'].astype(str)+'/'+df_booked.loc[:,'arrival_year'].astype(str), format='%d/%m/%Y')
```

```python
df_booked.head()
```

```python
df_booked.loc[:,'dias_pedido'] = df_booked.loc[:,'no_of_weekend_nights']+df_booked.loc[:,'no_of_week_nights']
df_booked.head()
```

```python
df_booked.loc[:, 'data_saida'] = df_booked['data_chegada']+ pd.to_timedelta(df_booked['dias_pedido'], unit='D')
```

Vamos entender a ocupação do hotel no início do dia.
- Quem sai no dia é contado e quem entra no dia só é contado no próximo

```python
import numpy as np

datetime_range = pd.date_range(start='2017-07-01', end='2018-12-31', freq='D')
data_chegada = df_booked.data_chegada.astype(str).values.tolist()
data_saida = df_booked.data_saida.values
dias_hosp = df_booked.dias_pedido.values
hospedes = np.zeros(len(datetime_range))

for i, chegada in enumerate(data_chegada):
    idx_chegada = datetime_range.get_loc(chegada)
    dias = dias_hosp[i]
    for j in range(dias):
        try:
            hospedes[idx_chegada+j+1] += 1
        except:
            continue
```

```python
import matplotlib.pyplot as plt

plt.plot(datetime_range, hospedes)
```

```python
df.loc[(df.arrival_month == 2) & (df.arrival_date == 29),'arrival_date'] = 28
df.loc[:, 'data_chegada'] = pd.to_datetime(df.loc[:,'arrival_date'].astype(str)+'/'+df.loc[:,'arrival_month'].astype(str)+'/'+df.loc[:,'arrival_year'].astype(str), format='%d/%m/%Y')
df.loc[:,'dias_pedido'] = df.loc[:,'no_of_weekend_nights']+df.loc[:,'no_of_week_nights']
df.loc[:, 'data_saida'] = df['data_chegada']+ pd.to_timedelta(df['dias_pedido'], unit='D')
datetime_range_all = pd.date_range(start='2017-07-01', end='2018-12-31', freq='D')
data_chegada_all = df.data_chegada.astype(str).values.tolist()
data_saida_all = df.data_saida.values
dias_hosp_all = df.dias_pedido.values
hospedes_all = np.zeros(len(datetime_range))

for i, chegada in enumerate(data_chegada_all):
    idx_chegada = datetime_range_all.get_loc(chegada)
    dias = dias_hosp_all[i]
    for j in range(dias):
        try:
            hospedes_all[idx_chegada+j+1] += 1
        except:
            continue
plt.plot(datetime_range_all, hospedes_all)
```

```python
line1, = plt.plot(datetime_range, hospedes_all, 'r--', label='Com desistências')
line2, = plt.plot(datetime_range, hospedes, 'b-', label='Sem desistências')
plt.xticks(fontsize=8)
plt.xlabel('data')
plt.ylabel('hospedes')
plt.legend(handles=[line1, line2])
plt.show()
```
